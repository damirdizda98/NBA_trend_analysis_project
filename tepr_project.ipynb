```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
nba = pd.read_csv("final.csv")
nba
```

```python
print(nba.columns)
```

```python
max_gp_by_season = nba.groupby('season')['G.x'].max()
max_gp_by_season
```

```python
# Remove the hyphen and everything after it in the Position column
nba['Pos.x'] = nba['Pos.x'].str.split('-').str[0]
```

```python
nba2 = nba.groupby('season').mean()[['player_height', 'player_weight']]
print(nba2)
```

```python
fig, ax1 = plt.subplots()

# Plot average height
ax1.plot(nba2['player_height'], color='red')

# Set y-axis label
ax1.set_ylabel('Average Height (cm)', color='red')

# Create second y-axis
ax2 = ax1.twinx()

# Plot average weight
ax2.plot(nba2['player_weight'], color='black')

# Set y-axis label
ax2.set_ylabel('Average Weight (kg)', color='black')

# Set plot title
plt.title('Average Height and Weight Trend From 1996-2022')

# Set tick locations and labels
plt.xticks(range(0, len(nba2.index), 4), nba2.index[range(0, len(nba2.index), 4)], rotation=45)

plt.show()
```

```python
mean_height_by_season_position = nba.groupby(['season', 'Pos.x'])[['player_height']].mean()

# Create a pivot table with season as rows, position as columns, and mean height as values
pivot_table = mean_height_by_season_position.pivot_table(index='season', columns='Pos.x', values='player_height')

# Plot the pivot table as a line chart
pivot_table.plot(kind='line')
plt.title('Mean Height of Players by Season and Position')
plt.xlabel('Season')
plt.ylabel('Mean Height (cm)')
plt.show()
```

```python
max_gp_season = nba.groupby('season')['G.x'].max()
nba1 = nba[nba['G.x'] >= 0.45 * nba['season'].map(max_gp_season)]
nba1
```

```python
top_20_players = nba1.groupby('season').apply(lambda x: x.nlargest(15, 'pts')).reset_index(level=0, drop=True).reset_index()

# Calculate the mean value of the points column for these players
mean_top_20_scorers = top_20_players.groupby('season')['X3PAr'].mean()
mean_top_20_scorers.plot(kind="line")

plt.title('Percentage of 3 point shots taken by top 20 scorers in each season')
plt.xlabel('Season')
plt.ylabel('Percentage of 3 point shots')

# Show the plot
plt.show()
```

```python
mean_top_20_scorers
```

```python
top_10_rebounders = nba1[nba1['Pos.x'] == 'PG'].groupby('season').apply(lambda x: x.nlargest(8, 'TRB.')).reset_index(level=0, drop=True).reset_index()

# Calculate the mean value of the points column for these players
top_10_rebounders = top_10_rebounders.groupby('season')['TRB.'].mean()

# Plot the results
top_10_rebounders.plot(kind="line")

plt.title('Percentage of rebounds recorded by top 10 PGs')
plt.xlabel('Season')
plt.ylabel('Percentage of rebounds')

# Show the plot
plt.show()
```

```python
top_10_rebounders = nba1[nba1['Pos.x'] == 'C'].groupby('season').apply(lambda x: x.nlargest(8, 'AST.')).reset_index(level=0, drop=True).reset_index()

# Calculate the mean value of the points column for these players
top_10_rebounders = top_10_rebounders.groupby('season')['AST.'].mean()

# Plot the results
top_10_rebounders.plot(kind="line")

plt.title('Percentage of assists recorded by top 10 Cs')
plt.xlabel('Season')
plt.ylabel('Percentage of assists')

# Show the plot
plt.show()
```
