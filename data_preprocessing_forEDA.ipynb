```python
import numpy as np
import pandas as pd
```

```python
nba = pd.read_csv("combined.csv")
nba
```

```python
print(nba.columns)
```

```python
nba = nba.drop(['Unnamed: 0', 'gp', 'Player.x',
       'player_name', 'PTS', 'TRB', 'AST', 'ts_pct', 'usg_pct',
               'oreb_pct', 'dreb_pct', 'ast_pct', 'team_abbreviation', 'net_rating', 'Tm.x', 'GS', 'BPM', 'VORP', 'WS', 'PER'], axis = 1)
```

```python
nba
```

```python
column_to_move = 'normalized_name'

cols = [column_to_move] + [col for col in nba if col != column_to_move]
nba = nba[cols]
```

```python
nba["Pos.x"] = nba["Pos.x"].str.split("-").str[0]
nba
```

```python
nba = nba.fillna(0)
```

```python
marcus_williams_data = nba[nba['normalized_name'] == 'Marcus Williams']
marcus_williams_data
```

```python
nba = nba.drop(index=[3377, 4006])

nba.reset_index(drop=True, inplace=True)
nba
```

```python
nba["country"] = nba["country"].replace("Serbia and Montenegro", "Serbia")
nba
```

```python
nba.to_csv('final_eda.csv', index=False)
```
